```python
import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://drive.google.com/uc?export=download&id=12KieiuP24jrrpTQqT4lj9L5iUXyr63s1')
```

```python
features_kept=['EDUCA', 'DEAF', 'EMTSUPRT', 'ADDEPEV3', 'SMOKE100', 'EXERANY2', '_BMI5CAT',
               'CVDINFR4', 'CVDCRHD4', 'CVDSTRK3', 'DIABETE4', 'CIMEMLOS']
df2 = df[features_kept]
```

```python
## EDUCA  - Limited education  Yes-1, No-0###


remove_idx = df2[df2.EDUCA == 9].index
df3 = df2.drop(index=remove_idx)
mapping = {1: 'Yes', 2: 'Yes', 3: 'No', 4: 'No', 5: 'No', 6: 'No'}
df3.EDUCA = df3.EDUCA.map(mapping)
mapping2 = {'Yes': 1, 'No': 0}
df3.EDUCA = df3.EDUCA.map(mapping2)
df3
```

```python
## DEAF  Yes-1, No-0###

remove_idx = df3[(df3.DEAF == 9) | (df3.DEAF == 7)].index
df4 = df3.drop(index=remove_idx)
df4.DEAF.value_counts()
df4.DEAF = df4.DEAF.astype(int)
df4.DEAF = df4.DEAF.replace(2, 0)
df4
```

```python
## EMTSUPRT Enough emotional support? Yes-1, No-0###

remove_idx = df4[(df4.EMTSUPRT == 9) | (df4.EMTSUPRT == 7)].index
df5 = df4.drop(index=remove_idx)
df5 = df5.dropna(subset = ['EMTSUPRT'])
df5.EMTSUPRT.value_counts()
mapping = {1: 'Yes', 2: 'Yes', 3: 'Yes', 4: 'No', 5: 'No'}
df5.EMTSUPRT = df5.EMTSUPRT.map(mapping)
mapping2 = {'Yes': 1, 'No': 0}
df5.EMTSUPRT = df5.EMTSUPRT.map(mapping2)
df5.EMTSUPRT = df5.EMTSUPRT.astype(int)
df5
```

```python
## ADDEPEV3 Depressive disorder? Yes-1, No-0

df5.ADDEPEV3.value_counts()
remove_idx = df5[(df5.ADDEPEV3 == 9) | (df5.ADDEPEV3 == 7)].index
df6 = df5.drop(index=remove_idx)

df6.ADDEPEV3 = df6.ADDEPEV3.astype(int)
df6.ADDEPEV3 = df6.ADDEPEV3.replace(2, 0)
df6
```

```python
## SMOKE100 smoked at least 100 cigarettes in your entire life?

df6.SMOKE100.value_counts()
remove_idx = df6[(df6.SMOKE100 == 9) | (df6.SMOKE100 == 7)].index
df7 = df6.drop(index=remove_idx)

df7.SMOKE100 = df7.SMOKE100.astype(int)
df7.SMOKE100 = df7.SMOKE100.replace(2, 0)
df7
```

```python
def binary_process(df, column):
  remove_idx = df[(df[column] == 9) | (df[column] == 7)].index
  df2 = df.dropna(subset = [column])
  df2 = df2.drop(index=remove_idx)
  df2[column] = df2[column].astype(int)
  df2[column] = df2[column].replace(2, 0)
  return df2
```

```python
## EXERANY2 During the past month, other than your regular job, did you participate in any
## physical activities or exercises such as running, calisthenics, golf, gardening, or walking for exercise?

df8 = binary_process(df7, 'EXERANY2')
df8
```

```python
## _BMI5CAT
remove_idx = df8[(df8['_BMI5CAT'] == 9) | (df8['_BMI5CAT'] == 7)].index
df9 = df8.dropna(subset = ['_BMI5CAT'])
df9 = df9.drop(index=remove_idx)
mapping = {1: 'No', 2: 'No', 3: 'No', 4: 'Yes'}
df9._BMI5CAT = df9._BMI5CAT.map(mapping)
mapping2 = {'Yes': 1, 'No': 0}
df9._BMI5CAT = df9._BMI5CAT.map(mapping2)
df9
```

```python
## CVDINFR4  (Ever told) you had a heart attack?
df10 = binary_process(df9, 'CVDINFR4')
df10
```

```python
## CVDCRHD4 coronary heart disease?
df11 = binary_process(df10, 'CVDCRHD4')
df11
```

```python
## CVDSTRK3  (Ever told) (you had) a stroke?
df12 = binary_process(df11, 'CVDSTRK3')
df12
```

```python
## DIABETE4   (Ever told) (you had) diabetes?
df12.DIABETE4.value_counts()
remove_idx = df12[(df12.DIABETE4 == 9) | (df12.DIABETE4 == 7)].index
df13 = df12.drop(index=remove_idx)
mapping = {1: 'Yes', 2: 'No', 3: 'No', 4: 'No'}
df13.DIABETE4 = df13.DIABETE4.map(mapping)
mapping2 = {'Yes': 1, 'No': 0}
df13.DIABETE4 = df13.DIABETE4.map(mapping2)
df13
```

```python
## CIMEMLOS
df14 = binary_process(df13, 'CIMEMLOS')
df14
```

```python
df14.reset_index(drop=True, inplace=True)
df14
```
