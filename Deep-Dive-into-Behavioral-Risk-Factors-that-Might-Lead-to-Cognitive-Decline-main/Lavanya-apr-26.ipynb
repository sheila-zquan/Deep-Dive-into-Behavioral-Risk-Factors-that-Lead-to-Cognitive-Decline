```python
subdf = df2[(df2.CIMEMLOS == 1)]
for column in ['CDHOUSE','CDASSIST', 'CDSOCIAL', 'CDDISCUS']:
  print(subdf[column].value_counts())
```

```python
drop_idx1 = df2[(df2.CIMEMLOS == 7) | (df2.CDHOUSE == 9)].index
drop_idx2 = subdf[(subdf.CDHOUSE == 7) | (subdf.CDHOUSE == 9) | (subdf.CDHOUSE.isna())].index
drop_idx3 = subdf[(subdf.CDASSIST == 7) | (subdf.CDASSIST == 9)| (subdf.CDASSIST.isna())].index
drop_idx4 = subdf[(subdf.CDSOCIAL == 7) | (subdf.CDSOCIAL == 9)| (subdf.CDSOCIAL.isna())].index
drop_idx5 = subdf[(subdf.CDDISCUS == 7) | (subdf.CDDISCUS == 9)| (subdf.CDDISCUS.isna())].index
```

```python
drop_idx = drop_idx1.union(drop_idx2).union(drop_idx3).union(drop_idx4).union(drop_idx5)
df3 = df2.drop(drop_idx)
```

```python
y_columns=['CIMEMLOS','CDHOUSE','CDASSIST','CDHELP','CDSOCIAL']
output = df[y_columns]
```

```python
refused_idx = output['CIMEMLOS'] == 9
remaining_output = output[~refused_idx]
```

```python
remaining_output.CIMEMLOS.replace(7, 1, inplace=True)
remaining_output.CIMEMLOS.replace(2, 0, inplace=True)
```

```python
mapping = {1: 'Always', 2: 'Usually', 3: 'Sometimes', 4: 'Rarely', 5: 'Never', 7: 'Never', 9: 'Never', 0: 'Missing'}
for column in ['CDHOUSE','CDASSIST','CDHELP','CDSOCIAL']:
  remaining_output[column] = remaining_output[column].fillna(0)
  remaining_output[column] = remaining_output[column].map(mapping) # Replace Don't know/not sure and Refused with Never
```

```python
mapping2 = {'Always': 5, 'Usually': 4, 'Sometimes': 3, 'Rarely': 2, 'Never': 1, 'Missing': 1}
for column in ['CDHOUSE','CDASSIST','CDHELP','CDSOCIAL']:
  remaining_output[column] = remaining_output[column].map(mapping2)
```

```python
remaining_output['y_sum'] = remaining_output[['CDHOUSE','CDASSIST','CDHELP','CDSOCIAL']].sum(axis=1) * remaining_output['CIMEMLOS']
```

```python
remaining_output['y_binary'] = remaining_output['y_sum'] > 0
remaining_output.y_binary = remaining_output.y_binary.astype(int)
```

```python
remaining_output.drop(columns=['CIMEMLOS', 'CDHOUSE','CDASSIST','CDHELP','CDSOCIAL'], inplace=True)
remaining_output
```

```python
numerical_columns = ['SLEPTIM1','WEIGHT2','HEIGHT3','CNCRAGE','_AGE80','_DRNKWK2']
df_numerical = df3[numerical_columns]
```

```python
df_numerical.CNCRAGE.fillna(0, inplace=True)
df_numerical.CNCRAGE.replace([98,99],0, inplace=True)
```

```python
mean_sleep = df_numerical[df_numerical['SLEPTIM1']<77].SLEPTIM1.mean()
df_numerical.SLEPTIM1.replace([77,99], mean_sleep, inplace=True)
```

```python
weight_sub = df_numerical[df_numerical.WEIGHT2 < 1000]
weight_sub['WEIGHT2'] = weight_sub['WEIGHT2'] * 0.453592
df_numerical.update(weight_sub)

weight_sub2 = df_numerical[(df_numerical.WEIGHT2 > 9000) & (df_numerical.WEIGHT2 < 9999)]
weight_sub2['WEIGHT2'] = (weight_sub2['WEIGHT2'] - 9000)
df_numerical.update(weight_sub2)

weight_sub3 = df_numerical[(df_numerical.WEIGHT2 < 7000)]
mean_weight = weight_sub3.WEIGHT2.mean()
df_numerical.WEIGHT2.replace([7777, 9999], mean_weight, inplace=True)
```

```python
height_sub = df_numerical[(9000 < df_numerical.HEIGHT3) & (df_numerical.HEIGHT3 < 9999)]
height_sub['HEIGHT3'] = (height_sub['HEIGHT3'] - 9000)/ 100
df_numerical.update(height_sub)

height_sub2 = df_numerical[df_numerical.HEIGHT3 < 1000]
height_sub2['HEIGHT3'] = np.floor(height_sub2['HEIGHT3']/100) * 0.3048 + (height_sub2['HEIGHT3']%100) * 0.0254
df_numerical.update(height_sub2)

height_sub3 = df_numerical[(df_numerical.HEIGHT3 < 7000)]
mean_height = height_sub3.HEIGHT3.mean()
df_numerical.HEIGHT3.replace([7777, 9999], mean_height, inplace=True)
```

```python
a=df_numerical[df_numerical['_DRNKWK2']<99900]
a['_DRNKWK2'] = a['_DRNKWK2']/100
df_numerical.update(a)

mean_drinks = a._DRNKWK2.mean()
df_numerical._DRNKWK2.replace(99900, mean_drinks, inplace=True)
```
