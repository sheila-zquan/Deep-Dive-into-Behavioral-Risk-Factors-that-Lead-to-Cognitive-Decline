## 2 to 0, one-hot dummy, /100

```python
# ['_TOTINDA', '_EXTETH3', '_ALTETH3', '_DENVST3', '_LTASTH1', '_CASTHM1', '_ASTHMS1', '_DRDXAR2', '_RACEG22',
#  '_AGE65YR','WTKG3', '_EDUCAG', '_INCOMG1', '_RFMAM22', '_SMOKER3', '_RFSMOK3','DRNKANY6', '_DRNKWK2', '_AIDTST4', 'Class']
```

```python
def ToBinary(df, column_name):
  df2 = df.copy()
  df2[column_name].replace(2.0, 0.0, inplace=True)
  return df2
```

```python
# Binary: '_EXTETH3', '_ALTETH3', '_DENVST3', '_LTASTH1', '_CASTHM1','_DRDXAR2', '_RACEG22','_RFMAM22','_RFSMOK3','DRNKANY6','_AIDTST4'
df = ToBinary(df, '_TOTINDA')
df = ToBinary(df, '_EXTETH3')
df = ToBinary(df, '_ALTETH3')
df = ToBinary(df, '_DENVST3')
df = ToBinary(df, '_LTASTH1')
df = ToBinary(df, '_CASTHM1')
df = ToBinary(df, '_DRDXAR2')
df = ToBinary(df, '_RACEG22')
df = ToBinary(df, '_RFMAM22')
df = ToBinary(df, '_RFSMOK3')
df = ToBinary(df, 'DRNKANY6')
df = ToBinary(df, '_AIDTST4')
```

```python
# '_AGE65YR': 1 to 0, 2 to 1
df['_AGE65YR'].replace(1.0, 0.0, inplace=True)
df['_AGE65YR'].replace(2.0, 1.0, inplace=True)
df
```

```python
# One-hot dummy: '_ASTHMS1', '_EDUCAG', '_SMOKER3'
df_ASTHMS1 = pd.get_dummies(df['_ASTHMS1'], prefix='_ASTHMS1_')
df_EDUCAG = pd.get_dummies(df['_EDUCAG'], prefix='_EDUCAG_')
df_SMOKER3 = pd.get_dummies(df['_SMOKER3'], prefix='_SMOKER3_')
```

```python
def subtract_100(df, col_name):
    df[col_name] = df[col_name]/100
    return df
```

```python
# Numerical: 'WTKG3','_DRNKWK2'
df = subtract_100(df, 'WTKG3')
df = subtract_100(df, '_DRNKWK2')
df
```

```python
# drop original and concat one-hot dummy cols
df.drop(columns=['_ASTHMS1', '_EDUCAG', '_SMOKER3'], inplace=True)
df2 = pd.concat([df_ASTHMS1, df_EDUCAG, df_SMOKER3], axis=1)
df = pd.concat([df, df2], axis=1)
df
```

## Impute

```python
def Impute(df, column_name, Unknown=7, Missing=9, Impute_based_on=None):
  df2 = df.copy()

  if Impute_based_on is None:
    mode = df2[column_name].mode()[0]
    df2[column_name].replace([Unknown, Missing], mode, inplace=True)

  else:
    NA_filter = df2[column_name].isin([Unknown, Missing])
    g = df2[~NA_filter].groupby(Impute_based_on)
    Mode = {group: g.get_group(group)[column_name].mode()[0] for group in g.groups}

    sub_na = df2[NA_filter]
    sub_na[column_name] = sub_na[Impute_based_on].map(Mode)
    df2.update(sub_na)

  return df2
```

```python
df2 = df.copy()
df2.drop(columns=['_ALTETH3'])
```

```python
df2 = Impute(df2, '_DRDXAR2')
```

```python
df2['_RFMAM22'].fillna(0.0, inplace=True)
```

```python
df2['WTKG3'].fillna(df2['WTKG3'].mean(), inplace=True)
```
