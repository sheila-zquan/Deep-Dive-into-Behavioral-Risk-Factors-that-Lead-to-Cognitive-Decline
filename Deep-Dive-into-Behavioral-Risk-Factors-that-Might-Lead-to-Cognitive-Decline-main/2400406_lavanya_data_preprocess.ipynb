```python
import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
df=pd.read_csv('https://drive.google.com/uc?export=download&id=12KieiuP24jrrpTQqT4lj9L5iUXyr63s1')
```

```python
features_kept=['_STATE','SEXVAR','GENHLTH','PHYSHLTH','MENTHLTH','PRIMINSR','PERSDOC3','CHECKUP1','EXERANY2','SLEPTIM1','CVDINFR4','CHCOCNC1','ADDEPEV3','CHCKDNY2','DIABETE4','MARITAL','VETERAN3','WEIGHT2','HEIGHT3','DEAF','BLIND','DECIDE','DIFFWALK','DIFFDRES','DIFFALON','COVIDPOS','CNCRDIFF','CNCRAGE','CNCRTYP2','CSRVPAIN','CIMEMLOS','CDHOUSE','CDASSIST','CDHELP','CDSOCIAL','LSATISFY','EMTSUPRT','SDHISOLT','EMPLOY1','SDHFOOD1','SDHBILLS','SDHUTILS','SDHTRNSP','SDHSTRE1','_URBSTAT','_IMPRACE','_RFHLTH','_PHYS14D','_MENT14D','_TOTINDA','_MICHD','_ASTHMS1','_DRDXAR2','_AGE80','_CHLDCNT','_EDUCAG','_INCOMG1','_SMOKER3','_CURECI2','_DRNKWK2','_AIDTST4']
df2 = df[features_kept]
```

```python
df3 = df2[['DIFFDRES', 'DIFFALON', 'COVIDPOS', 'CNCRDIFF', 'CNCRTYP2', 'LSATISFY', 'EMTSUPRT', 'SDHISOLT', 'EMPLOY1']]
```

```python
df3.head(100)
```

```python
for col in df3:
  print("col name is",col,"unique is",df3[col].unique())
```

```python
def ToBinary(df, column_name):
  df2 = df.copy()
  df2[column_name] = df2[column_name].astype(int)
  df2[column_name].replace(2, 0, inplace=True)
  return df2
```

```python
# col name is DIFFDRES unique is [2. 1. 9. 7.]
# col name is DIFFALON unique is [2. 1. 9. 7.]
# col name is COVIDPOS unique is [2. 1. 9. 3. 7.]

test = ToBinary(df3, 'DIFFDRES')
test = ToBinary(test, 'DIFFALON')
test = ToBinary(test, 'COVIDPOS')
```

```python
test['CNCRTYP2'].fillna(111, inplace=True)
test['LSATISFY'].fillna(111, inplace=True)
test['EMTSUPRT'].fillna(111, inplace=True)
test['SDHISOLT'].fillna(111, inplace=True)
test['CNCRDIFF'].fillna(7, inplace=True)

print("Unique values for LSATISFY column:")
print(test['LSATISFY'].unique())

print("Unique values for EMTSUPRT column:")
print(test['EMTSUPRT'].unique())

print("Unique values for SDHISOLT column:")
print(test['SDHISOLT'].unique())

print("Unique values for CNCRDIFF column:")
print(test['CNCRDIFF'].unique())
```

```python
cancer_mapping_dict = {
    1: 'Bladder',
    2: 'Blood',
    3: 'Bone',
    4: 'Brain',
    5: 'Breast',
    6: 'Cervix/Cervical',
    7: 'Colon',
    8: 'Esophagus/Esophageal',
    9: 'Gallbladder',
    10: 'Kidney',
    11: 'Larynx-trachea',
    12: 'Leukemia',
    13: 'Liver',
    14: 'Lung',
    15: 'Lymphoma',
    16: 'Melanoma',
    17: 'Mouth/tongue/lip',
    18: 'Ovary/Ovarian',
    19: 'Pancreas/Pancreatic',
    20: 'Prostate',
    21: 'Rectum/Rectal',
    22: 'Skin (non-melanoma)',
    23: 'Skin (don’t know what kind)',
    24: 'Soft tissue (muscle or fat)',
    25: 'Stomach',
    26: 'Testis/Testicular',
    27: 'Throat - pharynx',
    28: 'Thyroid',
    29: 'Uterus/Uterine',
    30: 'Other',
    77: 'Don’t know/Not sure',
    99: 'Refused',
    111: 'No Cancer'
}
test['CNCRTYP2'] = test['CNCRTYP2'].replace(cancer_mapping_dict)
```

```python
satisfaction_mapping = {
    1: 'Very satisfied',
    2: 'Satisfied',
    3: 'Dissatisfied',
    4: 'Very dissatisfied',
    7: 'Don’t know/Not sure',
    9: 'Refused',
    111: 'Missing'
}
test['LSATISFY'] = test['LSATISFY'].replace(satisfaction_mapping)
```

```python
employ_mapping_dict = {
    1: "Employed for wages",
    2: "Self-employed",
    3: "Out of work for 1 year or more",
    4: "Out of work for less than 1 year",
    5: "A homemaker",
    6: "A student",
    7: "Retired",
    8: "Unable to work",
    9: "Refused",
}

test['EMPLOY1'] = test['EMPLOY1'].replace(employ_mapping_dict)
```

```python
emotional_support_mapping = {
    1: 'Always',
    2: 'Usually',
    3: 'Sometimes',
    4: 'Rarely',
    5: 'Never',
    7: 'Don’t know/Not sure',
    9: 'Refused',
    111: 'Missing'
}
test['EMTSUPRT'] = test['EMTSUPRT'].replace(emotional_support_mapping)
```

```python
social_isolation_mapping = {
    1: 'Always',
    2: 'Usually',
    3: 'Sometimes',
    4: 'Rarely',
    5: 'Never',
    7: 'Don’t know/Not Sure',
    9: 'Refused',
    111: 'Missing'
}
test['SDHISOLT'] = test['SDHISOLT'].replace(social_isolation_mapping)
```

```python
cancer_type_mapping = {
    1: 'Only one',
    2: 'Two',
    3: 'Three or more',
    7: 'Don’t know/Not Sure',
    9: 'Refused',
}
test['CNCRDIFF'] = test['CNCRDIFF'].replace(cancer_type_mapping)
```

```python

one_hot_encoded = pd.get_dummies(test['CNCRTYP2'],dtype='int')
test = pd.concat([test, one_hot_encoded], axis=1)
test.drop(columns=['CNCRTYP2'], inplace=True)

one_hot_encoded = pd.get_dummies(test['LSATISFY'],dtype='int')
test = pd.concat([test, one_hot_encoded], axis=1)
test.drop(columns=['LSATISFY'], inplace=True)

one_hot_encoded = pd.get_dummies(test['EMPLOY1'],dtype='int')
test = pd.concat([test, one_hot_encoded], axis=1)
test.drop(columns=['EMPLOY1'], inplace=True)

one_hot_encoded = pd.get_dummies(test['EMTSUPRT'],dtype='int')
one_hot_encoded.columns = ['emotional_support_' + col.replace(' ', '_') for col in one_hot_encoded.columns]
test = pd.concat([test, one_hot_encoded], axis=1)
test.drop(columns=['EMTSUPRT'], inplace=True)

one_hot_encoded = pd.get_dummies(test['SDHISOLT'],dtype='int')
one_hot_encoded.columns = ['isolation_' + col.replace(' ', '_') for col in one_hot_encoded.columns]
test = pd.concat([test, one_hot_encoded], axis=1)
test.drop(columns=['SDHISOLT'], inplace=True)

one_hot_encoded = pd.get_dummies(test['CNCRDIFF'],dtype='int')
one_hot_encoded.columns = ['no_of_cancers_' + col.replace(' ', '_') for col in one_hot_encoded.columns]
test = pd.concat([test, one_hot_encoded], axis=1)
test.drop(columns=['CNCRDIFF'], inplace=True)
```
