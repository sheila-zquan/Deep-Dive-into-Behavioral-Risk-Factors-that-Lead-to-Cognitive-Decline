```python
df_new = pd.DataFrame()
```

```python
# (Ever told) (you had)  melanoma or any other types of cancer?
# Codebook: 1.0 = Yes, 2.0 = No
df2['CHCOCNC1']
df_new['CHCOCNC1'] = df2['CHCOCNC1'].replace(2.0, 0.0)

# if 7 or 9 (don't know/refused), change to 0 (No)
df_new.loc[~df_new['CHCOCNC1'].isin([1.0, 0.0]), 'CHCOCNC1'] = 0.0
df_new['CHCOCNC1'] = df_new['CHCOCNC1'].astype(int)
print(df_new.head())

unique_val = df_new['CHCOCNC1'].nunique()
print(unique_val)
```

```python
# (Ever told) you had a depressive disorder
# Codebook: 1.0 = Yes, 2.0 = No
df2['ADDEPEV3']
df_new['ADDEPEV3'] = df2['ADDEPEV3'].replace(2.0, 0.0)

# if 7 or 9 (don't know/refused), change to 0 (No)
df_new.loc[~df_new['ADDEPEV3'].isin([1.0, 0.0]), 'ADDEPEV3'] = 0.0
df_new['ADDEPEV3'] = df_new['ADDEPEV3'].astype(int)
print(df_new.head())

unique_val = df_new['ADDEPEV3'].nunique()
print(unique_val)
```

```python
# kidney disease
# Codebook: 1.0 = Yes, 2.0 = No
df2['CHCKDNY2']
df_new['CHCKDNY2'] = df2['CHCKDNY2'].replace(2.0, 0.0)

# if 7 or 9 (don't know/refused), change to 0 (No)
df_new.loc[~df_new['CHCKDNY2'].isin([1.0, 0.0]), 'CHCKDNY2'] = 0.0
df_new['CHCKDNY2'] = df_new['CHCKDNY2'].astype(int)
print(df_new.head())

unique_val = df_new['CHCKDNY2'].nunique()
print(unique_val)
```

```python
# Ever diabetes
# 1: Yes; 2: Yes, but female told only during pregnancy; 3:No;
# 4: No, pre-diabetes or borderline diabetes; 7: Don’t know/Not Sure; 9: Refused
df2['DIABETE4']
one_hot = pd.get_dummies(df2['DIABETE4'])
df_new2 = pd.DataFrame()
df_new2['DIABETE4'] = df2['DIABETE4']
df_new2 = pd.concat([df_new2, one_hot], axis=1)
df_new2.columns = ['DIABETE4'] + ['DIABETE4_' + str(col) for col in one_hot.columns]
df_new2

df_new = pd.concat([df_new, df_new2], axis=1)
df_new['DIABETE4'] = df_new['DIABETE4'].astype('category')

# drop original + 7.0 and 9.0
df_new = df_new.drop(columns=['DIABETE4', 'DIABETE4_7.0', 'DIABETE4_9.0'])

df_new
```

```python
# Marital status
# 1: Married; 2: Divorced; 3: Widowed; 4: Separated
# 5: Never married; 6: A member of an unmarried couple; 9: Refused
df2['MARITAL']
one_hot = pd.get_dummies(df2['MARITAL'])
df_new3 = pd.DataFrame()
df_new3['MARITAL'] = df2['MARITAL']
df_new3 = pd.concat([df_new3, one_hot], axis=1)
df_new3.columns = ['MARITAL'] + ['MARITAL_' + str(col) for col in one_hot.columns]
df_new3

df_new = pd.concat([df_new, df_new3], axis=1)
df_new['MARITAL'] = df_new['MARITAL'].astype('category')

# drop original + 9.0
df_new = df_new.drop(columns=['MARITAL', 'MARITAL_9.0'])
df_new
```
