```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
# from sklearn.linear_model import LinearRegression
# from sklearn.metrics import mean_squared_error, r2_score
# import seaborn as sns
```

```python
df=pd.read_csv('https://drive.google.com/uc?export=download&id=1sHR0Yor1dniuTP2U29vi7joMcZ_7tXxP')
```

```python
X = df.drop(columns=['y_sum', 'y_binary'])
y_binary = df.iloc[:, -1]
y_sum = df.iloc[:, -2]
```

```python
index = y_sum.index.tolist()
train_idx = np.random.choice(index, size=int(len(index) * 0.6), replace=False)
a = np.setdiff1d(index, train_idx)
val_idx = np.random.choice(a, size=int(len(a) * 0.5), replace=False)
test_idx = np.setdiff1d(a, val_idx)
```

```python
len(train_idx), len(val_idx), len(test_idx)
```

```python
X_train = X.iloc[train_idx]
X_val = X.iloc[val_idx]
X_test = X.iloc[test_idx]

y_binary_train = y_binary.iloc[train_idx]
y_binary_val = y_binary.iloc[val_idx]
y_binary_test = y_binary.iloc[test_idx]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
clf = RandomForestClassifier(n_estimators=100, class_weight='balanced')

# 训练模型
clf.fit(X_train, y_binary_train)

# 进行预测
y_pred = clf.predict(X_test)

```

```python
accuracy = accuracy_score(y_binary_test, y_pred)
```
